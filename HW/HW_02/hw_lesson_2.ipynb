```python
def get_vectorizer(VectorizerClass, max_df=0.9, max_features=1000, stop_words='english'):
    vectorizer = VectorizerClass(ngram_range=(1, 1), analyzer='word',
                                 max_df=max_df,
                                 max_features=max_features,
                                 stop_words=stop_words)

    bow = vectorizer.fit_transform(combine_df['tweet_stemmed'].apply(' '.join))
    bow = vectorizer.fit_transform(combine_df['tweet_lemmatized'].apply(' '.join))
    
    f_names = vectorizer.get_feature_names_out()
    bow_df = pd.DataFrame(bow.toarray(), columns=f_names)
    
    return bow_df, vectorizer
```

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import accuracy_score
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
def get_score(vectorizer, n_components=None):
    xtrain_vect = vectorizer.transform(train_x)
    xvalid_vect = vectorizer.transform(valid_x)
    
    if n_components:
        svd = TruncatedSVD(n_components=n_components, random_state=42)
        xtrain_vect = svd.fit_transform(xtrain_vect)
        xvalid_vect = svd.fit_transform(xvalid_vect)
    
    classifier = linear_model.LogisticRegression(max_iter=1000)
    classifier.fit(xtrain_vect, train_y)
    predictions = classifier.predict(xvalid_vect)
    return accuracy_score(valid_y, predictions)
```

```python
max_df_list = [1.0, 0.9, 0.8, 0.7]
max_features_list = [3000, 1000, 500, 250]
vec_class_list = [CountVectorizer, TfidfVectorizer]
n_components_list = [None, 5, 10, 20]

score_list = []

for max_df in max_df_list:
    for max_features in max_features_list:
        for VectorizerClass in vec_class_list:
            for n_components in n_components_list:
                bow, vectorizer = get_vectorizer(VectorizerClass, max_df=max_df, max_features=max_features)
                score = get_score(vectorizer, n_components=n_components)
                score_list.append({
                    'vec_class': VectorizerClass.__name__,
                    'max_df': max_df,
                    'max_features': max_features,
                    'n_components': n_components,
                    'score': score
                })
                print(score_list[-1])
```

```python
results = pd.DataFrame(score_list)
results.sort_values(by='score', ascending=False).head(15)
```

__ВЫВОДЫ:__ При равных значениях параметров max_df, max_features, n_components у TfidfVectorizer точность выше чем у CountVectorizer. При увеличении количества фичей (max_features) точность обоих векторайзеров растёт. Изменение значения параметра max_df от 0.7 до 1 особого влияния на точность векторайзеров не оказывает. При применении PCA точность снижается.
